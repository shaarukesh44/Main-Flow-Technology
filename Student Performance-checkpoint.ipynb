```python
# libraries
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Loading Dataset
dataset_path = "student-mat.csv"
data = pd.read_csv("student-mat.csv", sep=";")

# Step 2: DataLoad
print("Data Loaded Successfully!!")
print(data.head())
print("\nDataset Info:")
print(data.info())

# Step 3: Checking for missing values
print("\nMissing Values:")
print(data.isnull().sum())
#removing duplicates
data = data.drop_duplicates()

# Step 4: Data Analysis Questions
#Question 1: What is the average score in math(G3)?
average_score = data['G3'].mean()
print(f"Average Math Score (G3) : {average_score:.2f}")

#Question 2: How many student scored above 15 in their final grade (G3)?
above_15 = len(data[data['G3'] > 15])
print(f"Number of students scoring above 15:{above_15}")

#Question 3: Is there a correlation between study time(study time) and the final grade(G3)?
correlation = data['studytime'].corr(data['G3'])
print(f"Correlation between study time and final grade:{correlation:.2f}")

#Question 4: Which gender has a higher average final grade(G3)?
avg_by_gender = data.groupby('sex')['G3'].mean()
print("\nAverage Final Grade by Gender:")
print(avg_by_gender)

# Step 5: Data Visualization
#Histogram of final grades
plt.figure(figsize=(8,5))
plt.hist(data['G3'],bins=10, color='navy', edgecolor='white')
plt.title("Final Grades Distribution (G3)")
plt.xlabel("Final Grade")
plt.ylabel("Frequency")
plt.show()

#Scatter plot of study time and final grade
plt.figure(figsize=(8,5))
sns.scatterplot(data=data, x="studytime", y='G3',color=['purple'])
plt.title('Study Time vs Final Grade')
plt.xlabel("Study Time (hrs)")
plt.ylabel("Final Grade")
plt.legend(title="Gender")
plt.show()

#Bar chart of average scores by gender
plt.figure(figsize=(8,5))
avg_by_gender.plot(kind='bar', color=['pink','blue'])
plt.title("Average Final Grade by Gender")
plt.ylabel("Average Final Grade")
plt.xlabel("Gender")
plt.xticks(rotation=0)
plt.show()
```
